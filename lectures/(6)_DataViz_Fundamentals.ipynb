# Fundamentals of Data Visualization, LaTeX & Text Rendering, and Common Misconceptions



## Fundamentals of Data Visualization
Data visualization is the graphical representation of information and data. By using visual elements like charts, graphs, and maps, data visualization tools provide an accessible way to see and understand trends, outliers, and patterns in data.

**Advantages**
* Easily sharing information.
* Interactively explore opportunities.
* Visualize patterns and relationships.

**Disadvantages**
* Biased or inaccurate information.
* Correlation doesnâ€™t always mean causation.
* Core messages can get lost in translation.



### Plots




```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# forces random numbers to be the same each time
np.random.seed(42)

# sets 100 dots per in for resolution
plt.figure(dpi=100)

sample_size = 150
sample_mean = 0
sample_sd = 10
x1 = np.random.normal(sample_mean, sample_sd, sample_size)
x2 = np.linspace(-30,30,1000)
y1 = x1**2 + 2 * x1 + np.random.normal(0, 40, sample_size)
y2 = x2**2 + 2 * x2 + 40

# Creates plots
plt.plot(x1, y1, marker='.', linestyle='')
plt.plot(x2, y2)

# Displays plots
plt.show()
```

#### How can we add more information?

* Title
* Labels
* Grid
* Descriptive Text

```python
plt.plot(x1, y1, marker='.', linestyle='')
plt.plot(x2, y2)


plt.title("Relationship Example: y vs. x")
plt.xlabel("x (units)")
plt.ylabel("f(x) (units)")
plt.text(-15, 620, 'Equation: f(x) = x^2 + 2x - 5') #doesn't look too good
plt.grid(True)
plt.show()
```

Let's put it in MathText (LaTeX-like) by using $$

```python
plt.plot(x1, y1, marker='.', linestyle='')
plt.plot(x2, y2)

plt.title("Relationship Example: y vs. x")
plt.xlabel("$x$ (units)")
plt.ylabel("$f(x)$ (units)")

# In Matplotlib, the $ symbols tell it to interpret the enclosed text using LaTeX-style math formatting.
plt.text(-15, 620, 'Equation: $f(x) = x^2 + 2x - 5$')
plt.grid(True)
plt.show()
```

#### How can we make this look even nicer?

* Change font
* Add Colors
* ...More style parameters

```python
# Let's see what fonts we have natively supported
import matplotlib.font_manager as fm
print([f.name for f in fm.fontManager.ttflist])
```

```python
plt.plot(x1, y1, marker='.', linestyle='')
plt.plot(x2, y2)

# We can change things like font at the text element level
plt.text(-15, 620, 'Equation: $f(x) = x^2 + 2x - 5$', fontname="cmmi10")
plt.title("Relationship Example: y vs. x")
plt.xlabel("$x$ (units)")
plt.ylabel("$f(x)$ (units)")
plt.grid(True)
plt.show()
```

```python
# or we can change it globally

# Computer Modern (cm) is similar to LaTeX font
# rcParams is the style/config dictionary for all plots: fonts, figure size, colors, DPI, line width, etc
plt.rcParams.update({
    "font.family": "serif",
    "mathtext.fontset": "cm",
})

plt.plot(x1, y1, marker='.', linestyle='')
plt.plot(x2, y2)

plt.text(-15, 620, 'Equation: $f(x) = x^2 + 2x - 5$')
plt.title("Relationship Example: y vs. x")
plt.xlabel("$x$ (units)")
plt.ylabel("$f(x)$ (units)")
plt.grid(True)
plt.show()
```

What other parameters can we change? [Use the API Reference](https://matplotlib.org/stable/users/explain/customizing.html)!!!

```python
# Plot
plt.plot(
    x1, y1,
    marker='o',             # marker type
    markersize=6,           # marker size
    markerfacecolor='#EEB111', # marker fill color
    markeredgewidth=1.5,    # marker border width
    markeredgecolor='#002855', # marker border color
    linestyle='',           # line style
    alpha=0.8,              # transparency
    label='Points',         # legend label
)

# Plot text, title, and labels
plt.text(-10, 400,'Equation: $f(x) = x^2 + 2x - 5$',
    fontsize=12, # font size
    color='#002855',
    alpha=0.9,
    rotation=25, # display angle
    ha='left',  # horizontal alignment
    va='center', # vertical alignment
)

plt.title("Relationship Example: y vs. x",
    fontsize=18,
    fontstyle='italic',
    fontweight='bold',
    color='#002855',
    loc='center',
    pad=10
)

plt.xlabel(
    "$x$ (units)",
    fontsize=14,
    color='#002855',
    labelpad=10
)
plt.ylabel(
    "$f(x)$ (units)",
    fontsize=14,
    color='#002855',
    labelpad=10
)

# Grid and legend
plt.grid(True,
         which='both',
         linestyle='--',
         linewidth=0.7,
         color='gray',
         alpha=0.5)

plt.legend(fontsize=12,
           loc='upper center',
           facecolor='lightgray',
           edgecolor='black')

plt.plot(x2, y2)

plt.show()
```

```python
# The Drunkards Walk

# Walking home, taking 500 steps
t = range(0, 500)

each_step = np.random.normal(0, 1, 500)
distance_from_bar = np.cumsum(each_step)

plt.plot(t, distance_from_bar)
plt.title("Where is our Friend? Using plt")
plt.xlabel("Steps")
plt.ylabel("Location")
plt.show()
```

```python
# LinePlot via Seaborn
import pandas as pd
import seaborn as sns

# Seaborn is a Python data visualization library built on top of Matplotlib.
# Itâ€™s designed to make creating statistical graphics easier and prettier.
# Supports dataframes, prepackaged stylings, confidence intervals, etc.

# Seaborn utilizes data in DataFrame form
# Let's put steps and disance from bar into a df using dictionaries
df = pd.DataFrame({"Steps": t, "Location": distance_from_bar})

# Seaborn stylings
'''sns.set_style("darkgrid")  # grid background
sns.set_palette("pastel")   # color palette
sns.set_context("talk")     # makes fonts bigger for presentations'''
# sns.reset_defaults()

sns.lineplot(x="Steps", y="Location", data=df)
plt.title("Where is our Friend? Using Seaborn")
plt.show()
```


### Histograms


```python
# Making the Car Sale

# It's reasonable to use a gamma distribution to model time between sales
# The Gamma distribution is often used to model waiting times or non-negative skewed data.

customers_per_car_sold = 7 #
min_between_customers = 45
sales_per_year = 1000

car_data = np.random.gamma(shape=customers_per_car_sold,
                       scale=min_between_customers,
                       size=sales_per_year)

print(car_data[:5])

plt.hist(car_data, bins=20, edgecolor='black')
plt.title("Average Time Between Car Sales")
plt.xlabel("Minutes")
plt.ylabel("Sales per Bin")
plt.show()
```

### Heatmaps

```python
# Histoplot via Seaborn
# ----------------------------

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Column names
col_names = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width', 'Species']

# Load the dataset from UCI repository
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = pd.read_csv(url, header=None, names=col_names)
print(iris.head(5))
# Select only numeric columns using .select_dtypes
numeric_cols = iris.select_dtypes(include='number')
print(numeric_cols.head(5))

# Compute correlation matrix (r value from stats)
# https://en.wikipedia.org/wiki/Pearson_correlation_coefficient

# .corr() is a DataFrame method in pandas that calculates pairwise correlation
corr_matrix = numeric_cols.corr()

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix,
            annot=True,
            cmap='coolwarm',
            fmt='.2f',
            linewidths=0.5,
            vmin=-1,
            vmax=1 )
plt.title('Correlation Matrix of Iris Dataset')
plt.show()
```


## Visual Checklist
- Is the **question** clear from the title/subtitle?
- Are encodings **appropriate** for the data type?
- Are axes **labeled**, units present, legends minimal, and annotations helpful?
- Any **misleading scales**, truncated baselines, or overloaded color?
- Can a reader **get the key takeaway in 5 seconds**?



## Common Misconceptions & Pitfalls (with Fixes)

1. Don't use truncated axes
2. Reduce chartjunk; emphasize data, not effects.
3. Use labels/annotations; consider patterns/position for redundancy.
4. 3D distorts perception; use 2D unless spatial is the data.



### Truncated vs. zero-baseline bars


```python

import numpy as np
import matplotlib.pyplot as plt

cats = ["A","B","C","D"]
vals = np.array([48, 50, 52, 54])

# Misleading view (truncated)
plt.figure(figsize=(6,4))
plt.bar(cats, vals)
plt.ylim(47, 55)  # truncation exaggerates small differences
plt.title("Misleading: Truncated y-axis exaggerates differences")
plt.ylabel("Value")
plt.tight_layout()
plt.show()

# Proper baseline
plt.figure(figsize=(6,4))
plt.bar(cats, vals)
plt.ylim(0, 55)   # zero baseline
plt.title("Better: Bars with zero baseline")
plt.ylabel("Value")
plt.tight_layout()
plt.show()
```
